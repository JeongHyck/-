```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#데이터 불러오기
dia = datasets.load_diabetes()
#print(dia.DESCR)
print(dia.feature_names)
print(dia.data.shape)

df = pd.DataFrame(dia.data, columns=dia.feature_names)
df.head()
```

```python
#데이터 불러오기

df = pd.DataFrame(dia.data, columns=dia.feature_names)
df.head()
```

```python
#데이터 처리

dia_X = df['bmi'].values.reshape(-1,1)
print(dia_X.shape)
print(dia_X)
```

```python
#학습데이터 검증데이터 분활

dia_X_train = dia_X[:-20]
dia_X_test = dia_X[-20:]
print(dia_X_train.shape, dia_X_test.shape)

dia_y_train = dia.target[:-20]
dia_y_test = dia.target[-20:]

print(dia_y_train.shape, dia_y_test.shape)
```

```python
#학습

regr = linear_model.LinearRegression()

regr.fit(dia_X_train,dia_y_train)

print(regr.coef_)
print(regr.intercept_)
```

```python
#예측
y_pred = regr.predict(dia_X_test)
print(y_pred)
```

```python
#시각화

plt.scatter(dia_X_test, dia_y_test,label='True_value')
plt.plot(dia_X_test, y_pred, color = 'r', label = 'predicted')
plt.xlabel('bmi')
plt.ylabel("Progress")
plt.legend()
```

```python
#r2 스코어
print(r2_score(dia_y_test, y_pred))

#평균 제곱 오차
mean_squared_error(dia_y_test,y_pred)
```
