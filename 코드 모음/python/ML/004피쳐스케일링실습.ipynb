```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
X = np.arange(-3,6).astype("float32").reshape(-1,1)
X = np.vstack([X,[20]]) #가상의 이상치
print(X.shape)
print(X)
```

```python
#심플스케일링 simple scaling >>> 이미지 처리 0~255

X_simple = X/X.max()
X_simple
```

```python
#Min Max 스케일링
X_minmax = (X-X.min()) / (X.max()-X.min())
X_minmax
```

```python
sc = MinMaxScaler()
X_minmax2 = sc.fit_transform(X)
X_minmax2
```

```python
#standard 스케일링
X_standard = (X - X.mean()) / X.std()
X_standard
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,4,1)
plt.hist(X, bins=50)
plt.title('oringinal data')

plt.figure(figsize=(14,4))
plt.subplot(1,4,2)
plt.hist(X_simple, bins=50)
plt.title('simple scaling')

plt.figure(figsize=(14,4))
plt.subplot(1,4,3)
plt.hist(X_minmax, bins=50)
plt.title('minmax scaling')

plt.figure(figsize=(14,4))
plt.subplot(1,4,4)
plt.hist(X_standard, bins=50)
plt.title('standard scaling')
```
