```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#한국어 사용시 (맥)
import platform
from matplotlib import font_manager

if platform.system() == "Darwin":
    plt.rc('font', family = 'AppleGothic')

plt.rcParams['axes.unicode_minus'] = False
```

함수형 프로그래밍

```python
#점찍기
plt.plot(3, 2, "o",c='r')
plt.show()
```

```python
#선긋기
plt.plot([1,2,3],[4,5,1])
```

```python
#타이틀, 라벨 추가

x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

plt.plot(x,y)
plt.plot(x2,y2)
plt.title('test plot')
plt.xlabel("month")
plt.ylabel("price")
plt.show()
```

```python
#bar chart

x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

plt.bar(x,y,label="Korea")
plt.bar(x2,y2,label="US")

plt.xlabel('age')
plt.ylabel('person number')
plt.legend()
```

```python
#산점도

plt.scatter(x,y,label='train')
plt.scatter(x2,y2,label='test')
plt.legend()
```

```python
#히스토그램

np.random.seed(0)
x = np.random.randn(1000) #정규분포 난수 생성
plt.hist(x, bins = 10)
plt.ylabel('frequency')
plt.xlabel('random numper')
plt.show()
```

객체지향 프로그래밍 스타일

```python
import random

x = np.random.randint(low = 1, high = 11, size = 50)
y = x+np.random.randint(1, 5, size = x.size)

data = x+y
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(8,4))

ax1.scatter(x,y,c='r',marker='o',edgecolors='k')
ax1.set_title('SCATTER PLOT')
ax1.set_xlabel("X axis")
ax1.set_ylabel("Y axis")

ax2.hist(data, bins=20)
ax2.set_title("HISTOGRAM")
plt.show()
```
